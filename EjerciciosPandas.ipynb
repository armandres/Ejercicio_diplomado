```python
#Creamos una lista vacía sobre la cual vamos a ir añadiendo dataframes, 
#y a continuación escribimos un loop para ir leyendo los ficheros que 
#tienen el patrón generado anteriormente.

import pandas as pd
import glob

list_data = []
csv_files = glob.glob('*.csv')

# Escribimos un loop que irá a través de cada uno de los nombres de archivo a través 
#de globbing y el resultado final será la lista dataframes

for names in csv_files:
    data = pd.read_csv(names)
    list_data.append(data)
    
#Para chequear que todo está bien, mostramos la list_data por consola
list_data

#pd.concat(list_data,ignore_index=True)
```

```python
#Creación de un dataframe a partir de listas, primera parte
'''
Deseo crear un dataframe con los nombres de columna:
Nombre, grado y gastos.
En la primera parte creo listas
'''
nombres = ['Andrés', 'María Elena', 'Ali']
grados = ['Dr', 'Dr', 'Bachiller']
gastos = [1200, 1200, 3400]

habitantes = list(zip(nombres, grados, gastos))
habitantes
#df = pd.DataFrame(nombres, grados, gastos)
```

```python
#Creación de un dataframe a partir de listas, segunda parte
'''
En la segunda parte creo listas un dataframe
a partir de una lista de listas. Este objeto ya lo puedo
exportar como csv. Notar que las columnas no tienen los nombres
deseados. Aquí puedo agregar los nombres de las columnas diréctamente
al crear el df.
'''
hab_df = pd.DataFrame(habitantes, columns = ['Nombre', 'Grado', 'Gastos'])
hab_df
```

```python
#Creación de un dataframe a partir de listas, tercera parte
'''
Ahora agregaré los nombres correctos a cada columna
'''
hab2_df = pd.DataFrame(hab_df,
                            columns = ['Nombre', 'Grado', 'Gastos'])

hab2_df
```
